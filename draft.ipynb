```python


# Import necessary librariesimport numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.covariance import EllipticEnvelope

# Load the Iris dataset and select features
iris = load_iris(as_frame=True)
X = iris.data[['sepal length (cm)', 'sepal width (cm)']]

# Initialize the Elliptic Envelope model
model = EllipticEnvelope(contamination=0.05, support_fraction=0.8, random_state=42)

# Fit the model to the data
model.fit(X)

# Compute the anomaly scores for each point
scores = model.decision_function(X)

# Identify outliers based on the anomaly scores
outliers = np.where(scores < np.percentile(scores, 5))[0]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], color='green', label='Normal Data')
plt.scatter(X.iloc[outliers, 0], X.iloc[outliers, 1], color='red', label='Outliers')

plt.xlabel('Sepal Length (cm)', fontsize=13)
plt.ylabel('Sepal Width (cm)', fontsize=13)
plt.title('Anomaly Detection using Elliptic Envelope', fontsize=16)
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import OneClassSVM

# Load the Iris dataset and select features
iris = load_iris(as_frame=True)
X = iris.data[['sepal length (cm)', 'sepal width (cm)']]

# Initialize the One-Class SVM model
model = OneClassSVM(nu=0.05, kernel='rbf', gamma='scale')

# Fit the model to the data
model.fit(X)

# Compute the anomaly scores for each point
scores = model.decision_function(X)

# Identify outliers based on the anomaly scores
outliers = np.where(scores < np.percentile(scores, 5))[0]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], color='green', label='Normal Data')
plt.scatter(X.iloc[outliers, 0], X.iloc[outliers, 1], color='red', label='Outliers')

plt.xlabel('Sepal Length (cm)', fontsize=13)
plt.ylabel('Sepal Width (cm)', fontsize=13)
plt.title('Anomaly Detection using One-Class SVM', fontsize=16)
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor

# Load the Iris dataset and select features
iris = load_iris(as_frame=True)
X = iris.data[['sepal length (cm)', 'sepal width (cm)']]

# Initialize the Local Outlier Factor model
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

# Fit the model and calculate the outlier scores
outlier_labels = lof.fit_predict(X)
scores = lof.negative_outlier_factor_

# Identify the indices of the outliers
outliers = np.where(outlier_labels == -1)[0]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], color='green', label='Normal Data')
plt.scatter(X.iloc[outliers, 0], X.iloc[outliers, 1], color='red', label='Outliers')

plt.xlabel('Sepal Length (cm)', fontsize=13)
plt.ylabel('Sepal Width (cm)', fontsize=13)
plt.title('Anomaly Detection using Local Outlier Factor', fontsize=16)
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import IsolationForest

# Load the Iris dataset and select features
iris = load_iris(as_frame=True)
X = iris.data[['sepal length (cm)', 'sepal width (cm)']]

# Initialize the Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)

# Fit the model to the data
model.fit(X)

# Compute the anomaly scores for each point
scores = model.decision_function(X)

# Identify outliers based on the anomaly scores
outliers = np.where(scores < np.percentile(scores, 5))[0]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], color='green', label='Normal Data')
plt.scatter(X.iloc[outliers, 0], X.iloc[outliers, 1], color='red', label='Outliers')

plt.xlabel('Sepal Length (cm)', fontsize=13)
plt.ylabel('Sepal Width (cm)', fontsize=13)
plt.title('Anomaly Detection using Isolation Forest', fontsize=16)
plt.legend()
plt.show()
```
